```python
# copy F1_scaled.png in each folder in images folder to new folder named F1_scaled_images
import os
path = 'images'
new_path = 'F1_scaled_images'
if not os.path.exists(new_path):
    os.makedirs(new_path)

```

```python
import shutil

for folder in os.listdir(path):
    folder_path = os.path.join(path, folder)
    if os.path.isdir(folder_path):
        image_path = os.path.join(folder_path, 'F1_scaled.png')
        if os.path.exists(image_path):
            new_image_path = os.path.join(new_path, f'{folder}_F1_scaled.png')
            shutil.copy(image_path, new_image_path)
            print(f'Copied {image_path} to {new_image_path}')
        else:
            print(f'{image_path} does not exist')
```
